**Overview**

This exercise takes us to the earlier stages of the humanistic data query, namely the design and implementation of a model intended to represent an underlying source (in this case, a photographic collection) for the purposes of answering any queries we might have about this data. Think of this as both exploratory and hypothesis-testing: we'll come up with a generalized idea of how we will model the structured representation of our sources based on some preliminary questions we would have about the photographs' contents; but we will also design that data model so as not to foreclose additional analysis or new questions as we explore the data closely.

**Data Munging Challenge Time!**

Let's put our text wrangling and OpenRefine skills to practice!

Examining the list of entries, do we see a natural point that we can use to break the entries into columns? What columns do you foresee? Any other data integrity problems?

As it turns out, we can perform most of this split in OpenRefine. Let's give this a go:

1. Save your pasted text as a .txt file somewhere on your system where you can find it.

2. Launch OpenRefine and create a new project by browsing to the location where you saved the text file.

3. After clicking "Next," you may see a commonly occuring problem: by default, your OpenRefine preview may not have detected the encoding of the uploaded text file and you'll need to set it explicitly:

<img align="left" width="500" height="500" src="screencaptures/orefine_mojibake.png" alt="Mojibake Example"><br clear="both"><br/>

Click in the "Character encoding" box and in the popup, select UTF-8. In the preview, the accented characters should now be set correctly. Set a project name if you wish and click "Create Project"

4. We'll first slice out the creator/photgrapher. In the single column created, select "Edit column >> split into several columns." Split on the period (.) separator, and allow at most only 2 columns. In the resulting two columns, rename the first "photographer" and the second "description".

5. Now we want to retrive the year from the description column. We notice that we seem to have consistently rendered 4-digit years near the end of our entry. We'll use a special match function. In the dropdown for the newly created description column, select "Edit column >> Add column based on this column"

6. In the resulting window, give the new column the name "year" and in the expression box enter:
<pre>value.match(/.*(\d\d\d\d).*/)[0]</pre>

What does this expression ask OpenRefine to do? It says to look at the current value in the each cell, and look for ("match") any pattern that includes a string of 0 or more characters followed by 4 numerical digits and then 0 or more further characters. The result of this is a list of results, from which we take the first such match ( "[0]" )

In the preview, you should see our years appear. Click OK. We should now have three columns. Perform some last cleanup using text facets.

### 3. Geocoding (aka Geolocating)

Let's look at the spatial aspects of our records. One common and logical starting point is to establish single-point locations for each record in this table. Most mapping visualization systems rely on decimal-degree versions of Cartesian map coordinates -- in other words, the number of degrees from the equator (latitude) and number of degrees from the Prime Meridian (longitude), converted from degrees/minutes/seconds to a decimal number. Positive latitude or longitude indicate northern and easterly directions from 0 degrees, and negative indicate westerly and southern.

A fast way to find this is to use Google Maps. Let's look for as many subjects of our pictures as we can find using the following approach:

1. Enter the street address or cross street description into https://maps.google.com. You may need to do some historical digging using helpers like Wikipedia if the structure no longer exists but a sense of the general location can be found.

2. When you are satisfied you have found the location, right-click on the map at the spot where the subject stands (or stood) and select “What’s Here?” <br clear="both">

<img align="left" width="500" height="500" src="screencaptures/whats-here-gmaps.png" alt="Google Maps What's Here Function"><br clear="both"><br/>

3. Google will supply a small pop-up window with the decimal coordinates of the location given with a hyperlink. Click on that hyperlink.<br clear="both">

<img align="left" width="500" height="500" src="screencaptures/decimal-degrees.png" alt="Google Maps Finding Decimal Degrees"><br clear="both"><br/>

4. This will bring the latitude/longitude up into an info window on the left. Now you can copy and paste the comma-separated coordinates into our new columns, which we can call "coordinates."

### 4. Visualization

A great way to make quick visualizations of digital humanities data, especially for short-term exploratory uses, is Stanford's [Humanities + Design](http://hdlab.stanford.edu/) Lab's [Palladio](https://hdlab.stanford.edu/palladio/).

To launch our visualization, click on the homepage above, select "Start," and copy and paste the data from our shared spreadsheet into the window provided. Click "Load," and we'll then adjust Palladio's interpretation of our data.

A few notes about Palladio:

1. It is especially intended for one-off quick visualizations, though you can create a direct link and share out your visualizations.

2. It has both mapping capabilities and tabular data exploration affordances, including those for image libraries.

3. As with any mapping or visualization software, you'll need to pay attention to the formatting and data types of records you upload. For example, Palladio's timeline feature only works if it recognizes you have dates in your data, and the dates are properly formatted. Similarly, spatial coordinates need to be formatted correctly without errors to avoid missing records.
